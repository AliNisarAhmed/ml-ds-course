# Introduction to Matplotlib

```python
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
plt.plot();
```

```python
plt.plot([1, 2, 3, 4]);
```

```python
x= [1, 2, 3, 4]
y = [11, 22, 33, 44]
plt.plot(x, y);
```

```python
# 1st method
fig = plt.figure() # creates a figure
ax = fig.add_subplot()
plt.show()
```

```python
# 2nd method

fig = plt.figure()
ax = fig.add_axes([1, 1, 1, 1])
ax.plot(x, y)
plt.show()
```

```python
# 3rd method (recommended)

fig, ax = plt.subplots()
ax.plot(x, [11, 14, 19, 30]);
```

# Matplotlib example workflow

```python
# 0. import matplotlib
    # Done above at the top
    
# 1. Prepare Data
x = [1, 2, 3, 4]
y = [11, 22, 33, 44]

# 2. Set up plot
fig, ax = plt.subplots(figsize=(5, 8)) # (width, height)

# 3. Plot data
ax.plot(x, y)

# 4. customize plot
ax.set(title="Simple plot",
      xlabel="x-axis",
      ylabel="y-axis")

# 5. save and show (you save the whole figure)
fig.savefig("./images/sample-plot.png")
```

# Making figures with numpy arrays

We want: 
* Line plot
* Scatter plot
* Bar plot
* Histogram
* Subplots

```python
# Create some data
x = np.linspace(0, 10, 100)
```

```python
# Plot the data and create a line plot
fix, ax = plt.subplots()
ax.plot(x, x ** 2);
```

```python
# Use same data to make scatter plot

fig, ax = plt.subplots()
ax.scatter(x, np.exp(x))
```

```python
# Another scatter plot
fig, ax = plt.subplots()
ax.scatter(x, np.sin(x));
```

```python
# Make a plot from dictionary

nut_butter_prices = {"Almond butter": 10,
                    "Peanut butter": 8,
                    "Cashew butter": 12}
fig, ax = plt.subplots()
ax.bar(nut_butter_prices.keys(), height=nut_butter_prices.values())
ax.set(title="Dan's Nut butter store",
       ylabel="Price ($)");
```

```python
fig, ax = plt.subplots()
ax.barh(list(nut_butter_prices.keys()), list(nut_butter_prices.values()));
```

```python
# Make some data for Histogram
x = np.random.randn(1000)
fix, ax = plt.subplots()
ax.hist(x);
```

### Two options for subplots

```python
# Subplot option 1
fix, ((ax1, ax2), (ax3, ax4)) = plt.subplots(nrows=2,
                                            ncols=2,
                                            figsize=(10,5))
ax1.plot(x, x / 2);
ax2.scatter(np.random.random(10), np.random.rand(10));
ax3.bar(nut_butter_prices.keys(), nut_butter_prices.values());
ax4.hist(np.random.randn(1000));
```

```python
# Subplots option 2
fig, ax = plt.subplots(nrows=2,
                      ncols=2,
                      figsize=(10, 5))
# Plot to each different index
ax[0, 0].plot(x, x/ 2);
ax[0, 1].plot(np.random.random(10), np.random.random(10));
ax[1, 0].bar(nut_butter_prices.keys(), nut_butter_prices.values());
ax[1, 1].hist(np.random.randn(1000));
```

## which one should you use (pplot vs matplotlib OO method)?

* When plotting something quick, it's okay to use pyplot method
* When plotting more advanced, used the OO methods

```python
heart_disease.head()
```

```python
over_50 = heart_disease[heart_disease["age"] > 50]
over_50.head()
```

```python
# PyPlot method

over_50.plot(kind="scatter",
            x="age",
            y="chol",
            c="target",
            figsize=(10,6));
```

```python
# OO method

fig, ax = plt.subplots(figsize=(10, 6))
over_50.plot(kind="scatter",
            x="age",
            y="chol",
            c="target",
            ax=ax);
# ax.set_xlim([45, 90])
```
